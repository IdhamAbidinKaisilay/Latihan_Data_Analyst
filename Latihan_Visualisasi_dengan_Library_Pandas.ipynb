# Latihan Pandas: Visualisasi Data

```python
# Import library and data
import pandas as pd
import matplotlib.pyplot as plt

# plt.xkcd()
plt.style.use('seaborn-v0_8')

# Silahkan ganti path dibawah sesuai dengan google drive kalian!
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 16: Pandas - Visualisasi Data/laptop_price.csv'
df = pd.read_csv(path)
df.sample(20)
```

## 1. Buatlah Distribusi Harga Laptop!

```python
# Ketikkan kode anda disini:
plt.figure(figsize=(8, 5))

plt.hist(df['Price (Euro)'], bins=40, label='Price (Euro)')
plt.legend()
plt.title('Distribusi Harga Laptop (Euro)')
plt.xlabel('Harga (Euro)')
plt.ylabel('Frekuency')

plt.show()
```

## 2. Buatlah Scatter Plot Berat (Weight) dan Harga (Price) Laptop!

```python
df.info()
```

```python
# Ketikan kode anda disini:
fig, ax = plt.subplots()
df.plot(kind='scatter', x='Weight (kg)', y='Price (Euro)', title='Harga Laptop Berdasarkan Beratnya', ax=ax)
```

```python
# Hasil ouput yang diinginkan:
```

## 3. Buatlah horizontal bar plot rata-rata harga laptop berdasarkan perusahaan (company)!

```python
# Ketikan kode anda disini:
fig, ax = plt.subplots(1,1, figsize=(8,5))

company_mean = df.groupby('Company')['Price (Euro)'].mean().sort_index(ascending=True)
company_mean.plot(kind='barh', ax=ax, color='#FCDA00')

plt.title('Rata-Rata Harga Laptop Setiap Perusahaan')
plt.xlabel('Rata-Rata Harga (Euro)')
plt.ylabel('Company')

plt.show()
```

## 4. Buatlah boxplot harga laptop berdasarkan tipe laptop (TypeName), gunakan subplots!

```python
df['TypeName'].unique()
```

```python
# Ketikan kode anda disini:
df_ultrabook = df[df['TypeName'].isin(['Ultrabook'])]
df_notebook = df[df['TypeName'].isin(['Notebook'])]
df_netbook = df[df['TypeName'].isin(['Netbook'])]
df_gaming = df[df['TypeName'].isin(['Gaming'])]
df_2in1 = df[df['TypeName'].isin(['2 in 1 Convertible'])]
df_workstation = df[df['TypeName'].isin(['Workstation'])]

fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1,6, figsize=(16.2,5))

df_ultrabook.plot(kind='box', y='Price (Euro)', ax=ax1, title='Boxplot Harga\nLaptop Ultrabook')
df_notebook.plot(kind='box', y='Price (Euro)', ax=ax2, title='Boxplot Harga\nLaptop Notebook')
df_netbook.plot(kind='box', y='Price (Euro)', ax=ax3, title='Boxplot Harga\nLaptop Netbook')
df_gaming.plot(kind='box', y='Price (Euro)', ax=ax4, title='Boxplot Harga\nLaptop Gaming')
df_2in1.plot(kind='box', y='Price (Euro)', ax=ax5, title='Boxplot Harga\nLaptop 2 in 1 Convertible')
df_workstation.plot(kind='box', y='Price (Euro)', ax=ax6, title='Boxplot Harga\nLaptop Workstation')


# Samakan batas Y di semua axis
for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.set_ylim(0, 7000)

plt.tight_layout()
plt.show()
```
